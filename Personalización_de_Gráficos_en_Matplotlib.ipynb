# Personalización de Gráficos en Matplotlib

```python
import matplotlib.pyplot as plt

#Configuramos el tamaño del grafico
plt.figure(figsize=(8, 6))

#Configuramos el tamaño del grafico
hours = [2,3,4,5,6,7,8]
exam_scores_student_1 = [55, 60, 65, 70, 75, 80, 85]
exam_scores_student_2 = [50, 58, 63, 69, 74, 78, 83]

#Creamos el grafico de dispersión de dos estudiantes
plt.scatter(hours, exam_scores_student_1, marker='o', color='green', label='Estudiante 1')
plt.scatter(hours, exam_scores_student_2, marker='s', color='red', label='Estudiante 2')

#Configuramos el titulo y los ejes
plt.title('Relación entre horas estudiadas y el puntaje de dos estudiantes')
plt.xlabel('Horas')
plt.ylabel('Puntaje')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Configuramos el tamaño del gráfico
plt.figure(figsize=(8, 6))

# Datos
hours = [2, 3, 4, 5, 6, 7, 8]
exam_scores_student_1 = [56, 62, 68, 75, 80, 87, 93]
exam_scores_student_2 = [55, 60, 65, 70, 75, 80, 85]
exam_scores_student_3 = [50, 55, 60, 65, 70, 75, 80]
exam_scores_student_4 = [49, 50, 52, 55, 60, 70, 72]

# Creamos el gráfico de dispersión de cuatro estudiantes
plt.plot(hours, exam_scores_student_1, marker='o', color='blue', linestyle='-', linewidth=2, label='Estudiante 1')
plt.plot(hours, exam_scores_student_2, marker='s', color='purple', linestyle='--', linewidth=2, label='Estudiante 2')
plt.plot(hours, exam_scores_student_3, marker='^', color='pink', linestyle='-.', linewidth=2, label='Estudiante 3')
plt.plot(hours, exam_scores_student_4, marker='h', color='olive', linestyle=':', linewidth=2, label='Estudiante 4')

# Configuramos el título y los ejes
plt.title('Relación entre horas estudiadas y el puntaje de los estudiantes')
plt.xlabel('Horas')
plt.ylabel('Puntaje')

# Accedemos al área de dibujo (ejes actuales) y cambiamos el color de fondo
ax = plt.gca()  # Obtener los ejes actuales
ax.set_facecolor('lightblue')  # Cambiar el color del área de dibujo

# Mostrar la leyenda y el gráfico
plt.legend()
plt.show()
```
