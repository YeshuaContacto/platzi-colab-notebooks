# **Subplots y Layouts Avanzados**

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

x = np.linspace(0, 10, 100)
y = np.sin(x)

data = np.random.randn(100)
```

```python
gs = gridspec.GridSpec(2,2, height_ratios=[2,1],width_ratios=[1,1])
fig = plt.figure(figsize =(10,8))

# Primer subplot grande, ocupa toda la primera fila
ax1 = fig.add_subplot(gs[0,:])
ax1.plot(x,y, color = 'blue')
ax1.set_title('Seno de X')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')

# Segundo subplot, ocupa la esquina inferior izquierda
ax2 = fig.add_subplot(gs[1,0])
ax2.hist(data, bins = 20, color='purple', edgecolor='black')
ax1.set_title('Histograma')
ax1.set_xlabel('Valor')
ax1.set_ylabel('Frecuencia')

# Tercer subplot, ubicado en la esquian inferior derecha
ax3 = fig.add_subplot(gs[1,1])
ax3.scatter(x,y, color = 'red')
ax1.set_title('Dispersi√≥n de Seno')
ax1.set_xlabel('x')
ax1.set_ylabel('Frecuencia')

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

# Datos
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
data = np.random.randn(100)
data2 = np.random.randn(100)

# Configuraci√≥n del layout con GridSpec
gs = gridspec.GridSpec(3, 2, height_ratios=[2, 1, 1], width_ratios=[1, 1])
fig = plt.figure(figsize=(10, 10))

# Primer subplot grande, ocupa toda la primera fila
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(x, y, color='blue', label='sin(x)')
ax1.plot(x, y2, color='green', label='cos(x)')
ax1.set_title('Seno y Coseno de X')
ax1.set_xlabel('x')
ax1.set_ylabel('Valores')
ax1.legend()

# Segundo subplot, histograma en la esquina inferior izquierda
ax2 = fig.add_subplot(gs[1, 0])
ax2.hist(data, bins=20, color='purple', edgecolor='black')
ax2.set_title('Histograma de Datos Aleatorios')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Frecuencia')

# Tercer subplot, dispersi√≥n de seno en la esquina inferior derecha
ax3 = fig.add_subplot(gs[1, 1])
ax3.scatter(x, y, color='red')
ax3.set_title('Dispersi√≥n de Seno')
ax3.set_xlabel('x')
ax3.set_ylabel('sin(x)')

# Cuarto subplot, gr√°fico de barras en la parte inferior
ax4 = fig.add_subplot(gs[2, 0])
ax4.bar(np.arange(10), np.random.randint(1, 10, size=10), color='orange')
ax4.set_title('Gr√°fico de Barras')
ax4.set_xlabel('√çndice')
ax4.set_ylabel('Valor')

# Quinto subplot, gr√°fico de l√≠neas con ruido en la parte inferior derecha
ax5 = fig.add_subplot(gs[2, 1])
ax5.plot(x, y + np.random.randn(100)*0.1, color='brown')
ax5.set_title('Seno con Ruido')
ax5.set_xlabel('x')
ax5.set_ylabel('sin(x) + Ruido')

# Ajustar el layout
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

# Datos
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
data = np.random.randn(100)
data2 = np.random.randn(100)

# Colores profesionales
color_sin = '#1f77b4'   # Azul oscuro
color_cos = '#ff7f0e'   # Naranja
color_hist = '#2ca02c'  # Verde
color_scatter = '#9467bd'  # Morado
color_bar = '#d62728'   # Rojo oscuro
color_noise = '#17becf' # Cian

# Configuraci√≥n del layout con GridSpec
gs = gridspec.GridSpec(3, 2, height_ratios=[2, 1, 1], width_ratios=[1, 1])
fig = plt.figure(figsize=(10, 10))

# Primer subplot grande, ocupa toda la primera fila
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(x, y, color=color_sin, label='sin(x)', linewidth=2)
ax1.plot(x, y2, color=color_cos, label='cos(x)', linewidth=2, linestyle='--')
ax1.set_title('Seno y Coseno de X', fontsize=14, fontweight='bold')
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('Valores', fontsize=12)
ax1.legend()

# Segundo subplot, histograma en la esquina inferior izquierda
ax2 = fig.add_subplot(gs[1, 0])
ax2.hist(data, bins=20, color=color_hist, edgecolor='black')
ax2.set_title('Histograma de Datos Aleatorios', fontsize=14, fontweight='bold')
ax2.set_xlabel('Valor', fontsize=12)
ax2.set_ylabel('Frecuencia', fontsize=12)

# Tercer subplot, dispersi√≥n de seno en la esquina inferior derecha
ax3 = fig.add_subplot(gs[1, 1])
ax3.scatter(x, y, color=color_scatter, s=40, alpha=0.7)
ax3.set_title('Dispersi√≥n de Seno', fontsize=14, fontweight='bold')
ax3.set_xlabel('x', fontsize=12)
ax3.set_ylabel('sin(x)', fontsize=12)

# Cuarto subplot, gr√°fico de barras en la parte inferior izquierda
ax4 = fig.add_subplot(gs[2, 0])
ax4.bar(np.arange(10), np.random.randint(1, 10, size=10), color=color_bar)
ax4.set_title('Gr√°fico de Barras', fontsize=14, fontweight='bold')
ax4.set_xlabel('√çndice', fontsize=12)
ax4.set_ylabel('Valor', fontsize=12)

# Quinto subplot, gr√°fico de l√≠neas con ruido en la parte inferior derecha
ax5 = fig.add_subplot(gs[2, 1])
ax5.plot(x, y + np.random.randn(100)*0.1, color=color_noise, linewidth=2)
ax5.set_title('Seno con Ruido', fontsize=14, fontweight='bold')
ax5.set_xlabel('x', fontsize=12)
ax5.set_ylabel('sin(x) + Ruido', fontsize=12)

# Ajustar el layout
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

# Datos
x = np.linspace(0, 10, 100)
y = np.sin(x)
y2 = np.cos(x)
data = np.random.randn(100)
data2 = np.random.randn(100)

# Colores profesionales para modo oscuro
color_sin = '#76c7c0'   # Azul claro brillante
color_cos = '#ffa07a'   # Naranja claro
color_hist = '#98fb98'  # Verde claro
color_scatter = '#dda0dd'  # Morado claro
color_bar = '#ff6347'   # Rojo tomate
color_noise = '#00ced1' # Cian brillante

# Configuraci√≥n del layout con GridSpec
gs = gridspec.GridSpec(3, 2, height_ratios=[2, 1, 1], width_ratios=[1, 1])
fig = plt.figure(figsize=(10, 10))

# Configuraci√≥n del fondo oscuro
plt.style.use('dark_background')

# Primer subplot grande, ocupa toda la primera fila
ax1 = fig.add_subplot(gs[0, :])
ax1.plot(x, y, color=color_sin, label='sin(x)', linewidth=2)
ax1.plot(x, y2, color=color_cos, label='cos(x)', linewidth=2, linestyle='--')
ax1.set_title('Seno y Coseno de X', fontsize=14, fontweight='bold', color='white')
ax1.set_xlabel('x', fontsize=12, color='white')
ax1.set_ylabel('Valores', fontsize=12, color='white')
ax1.legend(facecolor='black', edgecolor='white')

# Segundo subplot, histograma en la esquina inferior izquierda
ax2 = fig.add_subplot(gs[1, 0])
ax2.hist(data, bins=20, color=color_hist, edgecolor='white')
ax2.set_title('Histograma de Datos Aleatorios', fontsize=14, fontweight='bold', color='white')
ax2.set_xlabel('Valor', fontsize=12, color='white')
ax2.set_ylabel('Frecuencia', fontsize=12, color='white')

# Tercer subplot, dispersi√≥n de seno en la esquina inferior derecha
ax3 = fig.add_subplot(gs[1, 1])
ax3.scatter(x, y, color=color_scatter, s=40, alpha=0.7)
ax3.set_title('Dispersi√≥n de Seno', fontsize=14, fontweight='bold', color='white')
ax3.set_xlabel('x', fontsize=12, color='white')
ax3.set_ylabel('sin(x)', fontsize=12, color='white')

# Cuarto subplot, gr√°fico de barras en la parte inferior izquierda
ax4 = fig.add_subplot(gs[2, 0])
ax4.bar(np.arange(10), np.random.randint(1, 10, size=10), color=color_bar)
ax4.set_title('Gr√°fico de Barras', fontsize=14, fontweight='bold', color='white')
ax4.set_xlabel('√çndice', fontsize=12, color='white')
ax4.set_ylabel('Valor', fontsize=12, color='white')

# Quinto subplot, gr√°fico de l√≠neas con ruido en la parte inferior derecha
ax5 = fig.add_subplot(gs[2, 1])
ax5.plot(x, y + np.random.randn(100)*0.1, color=color_noise, linewidth=2)
ax5.set_title('Seno con Ruido', fontsize=14, fontweight='bold', color='white')
ax5.set_xlabel('x', fontsize=12, color='white')
ax5.set_ylabel('sin(x) + Ruido', fontsize=12, color='white')

# Ajustar el layout
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```
