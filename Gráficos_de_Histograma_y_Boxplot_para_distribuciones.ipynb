# **Gráficos de Histograma y Boxplot para distribuciones**

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

ages = [np.random.normal(30, 5, 100),
        np.random.normal(40, 5, 100),
        np.random.normal(35, 5, 100)]

# print(ages)

plt.boxplot(ages, notch=True, patch_artist=True, vert=True,
            labels=['Grupo 1', 'Grupo 2', 'Grupo 3', ])

plt.title('Distribución de edades por grupo')
plt.xlabel('Grupo')
plt.ylabel('Edad (años)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar los datos
data = np.random.normal(170, 10, 200)

# Estilo de gráfico
plt.style.use('seaborn-darkgrid')

# Crear la figura con tamaño personalizado
plt.figure(figsize=(10, 6))

# Histograma con la densidad activada
sns.histplot(data, color='salmon', bins=20, kde=True, edgecolor='black', alpha=0.6)

# Títulos y etiquetas
plt.title('Distribución de altura', fontsize=16, fontweight='bold')
plt.xlabel('Altura (cm)', fontsize=14)
plt.ylabel('Densidad', fontsize=14)

# Agregar información de estadísticas al gráfico
media = np.mean(data)
desviacion = np.std(data)
plt.axvline(media, color='blue', linestyle='dashed', linewidth=2, label=f'Media: {media:.2f} cm')
plt.axvline(media + desviacion, color='green', linestyle='dashed', linewidth=1, label=f'+1 Std: {media + desviacion:.2f} cm')
plt.axvline(media - desviacion, color='green', linestyle='dashed', linewidth=1, label=f'-1 Std: {media - desviacion:.2f} cm')

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
# Genera datos aleatorios a partir de una distribución normal multivariante
# y crea un histograma 2D con 30 bins
import numpy as np
import matplotlib.pyplot as plt

# Datos aleatorios
np.random.seed(0)

# Media para los ejes x, y
mean = [4, 8]

# Matriz de covarianza
cov = [[1, 0.5], [0.5, 0.11]]

# Generamos 10,000 medidas aleatorias
x, y = np.random.multivariate_normal(mean, cov, 10000).T

# Títulos y etiquetas
plt.title('Rendimiento academico', fontsize=16, fontweight='bold')
plt.xlabel('Horas estudiadas', fontsize=14)
plt.ylabel('Calificaciones', fontsize=14)

# Histogram 2D
plt.hist2d(x, y, bins=30, cmap='Blues')
plt.colorbar()
plt.show()
```
