# **Series Temporales en Pandas**

```python
import pandas as pd

#Cargar el dataset desde el archivo CSV
file_path = '/content/online_retail.csv'
df = pd.read_csv(file_path)

#Mostrar las primeras filas para entender la estructura del dataset
print("Primeras filas del dataset original")
df.head()
```

```python
#Cargo la informacion del dataset para saber que tipo de dato es la columna InvoiceDate
df.info()
```

```python
# Convierte la columna 'InvoiceDate' a formato de fecha y hora, y luego muestra un resumen de la información del DataFrame
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df.info()
```

```python
# Elimina las filas que tienen valores nulos en la columna 'InvoiceDate' y actualiza el DataFrame original
df.dropna(subset=['InvoiceDate'], inplace=True)
```

```python
# Establece la columna 'InvoiceDate' como el índice del DataFrame
df.set_index('InvoiceDate', inplace=True)
```

```python
# Crea nuevas columnas 'Year', 'Month', 'Day', 'Weekday' y 'Hour' extrayendo la información del índice basado en 'InvoiceDate'
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
df['Weekdy'] = df.index.weekday
df['Hour'] = df.index.hour

print(df['Year'])
print(df['Hour'])
```

```python
# Filtra el DataFrame para obtener solo los datos del año 2011 y muestra las primeras filas
df_2011 = df.loc['2011']

df_2011.head()
```

```python
# Filtra el DataFrame para obtener solo los datos de diciembre de 2011 y muestra las primeras filas
df_2011_dec = df.loc['2011-12']

df_2011_dec.head()
```

```python
# Filtra el DataFrame para obtener los datos desde el 1 de diciembre de 2010 hasta el 15 de diciembre de 2010 y muestra las primeras filas
df_dec_range = df.loc['2010-12-01':'2010-12-15']

df_dec_range.head()
```

```python
# Genera un rango de fechas diarias desde el 1 de enero de 2024 hasta el 31 de diciembre de 2024
date_range_new = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
date_range_new
```

```python
# Crea un DataFrame con el rango de fechas generado y lo asigna a la columna 'Date', luego muestra las primeras filas
df_dates = pd.DataFrame(date_range_new, columns=['Date'])
df_dates.head()
```
