# **Gráficos de Barras y Diagramas de Pastel**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

categories = ['Producto A', 'Producto B', 'Producto C']
sales = [120,150,90]

#Creación del grafico de barras verticales
plt.bar(categories, sales, color='skyblue', label='Ventas mensauales')

#Anotación con flecha para detectar un punto específico
plt.annotate('Máximo de ventas', xy=('Producto B', 150), xytext=('Producto C', 160),
             arrowprops=dict(facecolor='black'))


plt.title('Ventas de productos en un mes')
plt.xlabel('Productos')
plt.ylabel('Ventas')

#plt.legend()

plt.show()
```

```python
plt.figure(figsize=(8,6))

categories = ['Producto A', 'Producto B', 'Producto C']
sales = [120,150,90]

#Creación del grafico de barras verticales
plt.barh(categories, sales, color='skyblue', label='Ventas mensauales')



plt.title('Ventas de productos en un mes')
plt.ylabel('Productos')
plt.xlabel('Ventas')

#plt.legend()

plt.show()
```

```python
#Configuramos el tamaño de la figura
plt.figure(figsize=(8,6))

#Insertamos datos de ejemplo: Participación de mercado por producto
products = ['Producto A', 'Producto B', 'Producto C']
market_share = [50, 35, 15]

#Creamos el diagrama de pastel
plt.pie(market_share, labels=products, autopct='%1.1f%%', startangle=140,
        colors=['gold', 'lightcoral', 'orange'])

#Título
plt.title('Participación de Mercado por Producto')

#Aseguramos que el gráfico sea un círculo
plt.axis('equal')

#Mostramos el Gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Semilla
np.random.seed(19680801)

# Cálculo de las barras polares de ángulo theta
N = 20
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
radii = 10 * np.random.rand(N)
width = np.pi / 4 * np.random.rand(N)
colors = plt.cm.viridis(radii / 10.)

# Gráfico de barras en eje polar
ax = plt.subplot(projection='polar')
ax.bar(theta, radii, width=width, bottom=0.0, color=colors, alpha=0.5)
plt.show()
```
